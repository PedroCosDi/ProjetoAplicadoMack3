```python
#importando as bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn import preprocessing
```

```python
#leitura dos dados
df = pd.read_csv ('Base_Interações.csv', low_memory = False, sep = ";")
print(df)
```

```python
#excluindo colunas desncessárias
df.drop('date', axis = 1, inplace = True)
```

```python
#Numero de linhas e de colunas
df.head()
```

```python
#filtrando somente as colunas necessarias e renomeando nome das variaveis

#Selecionando somente as variaveis que iremos utilizar
df = df[['user_id','recipe_id', 'rating', 'review']]
```

```python
#nome das colunas
df.columns
```

```python
#Novo numero de linhas e de colunas
df.shape
```

```python
#Verificando valores ausentes
df.isna().sum()
```

```python
df.info()
```

```python
#Preenchendo as linhas de dados ausentes
enulo = df.isnull()
enulo.head(100)
```

```python
#Quantidade de dados ausentes
ausentes = df.isnull().sum()
print(ausentes)
```

```python
#Porcentagem dos dados ausentes
ausentes_percentual = (df.isnull().sum() / len(df['user_id']))
print(ausentes_percentual)
```

```python
#Substituindo os dados ausentes
df['user_id'].fillna('No_user_id', inplace = True)
df['recipe_id'].fillna('No_recipe_id', inplace = True)
df['rating'].fillna('No_rating', inplace = True)
df['review']. fillna ('No_review', inplace = True)

df.head(100)
```

```python
#Novo DataFrame após substituição dos valores ausentes
df.shape
```

```python
#Novo percentual após substuição dos valores ausentes
ausentes_percentual = (df.isnull(). sum() / len(df['user_id']))
print(ausentes_percentual)
```
