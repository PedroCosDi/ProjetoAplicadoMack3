```python
#importando as bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn import preprocessing
```

```python
#leitura dos dados
df = pd.read_csv ('Base_Receitas.csv',low_memory = False, sep=";")
print(df)
```

```python
#excluindo colunas desnecessarias
df.drop('contributor_id', axis = 1, inplace = True)
df.drop('submitted', axis = 1, inplace = True)
df.drop('n_steps', axis = 1, inplace = True)
df.drop('description', axis = 1, inplace = True)
df.drop('n_ingredients', axis = 1, inplace = True)
```

```python
df.head()
```

```python
#filtrando somente as colunas necessarias e renomeando nome das variaveis

#selecionando somente as variaveis que iremos utilizar
df = df [['name', 'id', 'minutes', 'tags', 'nutrition', 'steps', 'ingredients']]

#renomeando as variaveis
df.rename (columns = {'name':'Nome_Receita', 'id': 'Id_Receita','minutes':'Tempo', 'tags': 'Tags', 'nutrition': 'Valor_Nutricional', 'steps': 'Etapas', 'ingredients':'Ingredientes'}, inplace = True)

#exibindo as primeiras linhas do arquivo tratado
df.head()
```

```python
#Preenchendo as linhas de dados ausentes
enulo = df.isnull()
enulo.head(100)
```

```python
#Quantidade de dados ausentes
ausentes = df.isnull().sum()
print(ausentes)
```

```python
#Porcentagem dos dados ausentes
ausentes_percentual = (df.isnull().sum() / len(df['Nome_Receita']))
print(ausentes_percentual)
```

```python
#Substituindo os dados ausentes parte 1
df['Nome_Receita'].fillna('Sem_Nome', inplace = True)
df['Tags'].fillna('Sem_Tags', inplace = True)
df['Etapas'].fillna('Sem_Etapas', inplace = True)
df['Ingredientes'].fillna('Sem_Ingredientes', inplace = True)
df['Id_Receita'].fillna('Sem_Id', inplace = True)
df['Tempo'].fillna('Sem_Tempo', inplace = True)
df['Valor_Nutricional'].fillna('Sem_Valor_Nutricional', inplace = True)

df.head(100)
```
